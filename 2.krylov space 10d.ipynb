```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.random.rand(10)
A = np.random.rand(10,10)
```

```python
x
```

```python
plt.rcParams["figure.figsize"] = (15,15)

fig, ax = plt.subplots(2,3)
krylov = []
cur = x
for i in range(0,6):
    cur = cur/np.linalg.norm(cur)
    ax[0,0].plot([0,cur[0]],[0, cur[1]])
    ax[0,1].plot([0,cur[2]],[0, cur[3]])
    ax[1,0].plot([0,cur[4]],[0, cur[5]])
    ax[1,1].plot([0,cur[6]],[0, cur[7]])
    ax[1,2].plot([0,cur[8]],[0, cur[9]])
    krylov.append(cur)
    ax[0,0].annotate(i, xy = cur[0:2])
    ax[0,1].annotate(i, xy = cur[2:4])
    ax[1,0].annotate(i, xy = cur[4:6])
    ax[1,1].annotate(i, xy = cur[6:8])
    ax[1,2].annotate(i, xy = cur[8:10])
    cur = A.dot(cur)
plt.show()
```

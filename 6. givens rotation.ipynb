```python
phi=np.pi*9/10
s = np.sin(phi)
c = np.cos(phi)
F = np.array([[-c,s],[s,c]])
J = np.array([[c,s],[-s,c]])
x = np.array([1,0])
```

```python
fx = np.dot(F,x)
plt.plot([0,x[0]],[0,x[1]])
plt.annotate('x',x)
plt.plot([0,fx[0]],[0,fx[1]])
plt.annotate('fx',fx)
```

```python
jx = np.dot(J,x)
plt.plot([0,x[0]],[0,x[1]])
plt.annotate('x',x)
plt.plot([0,jx[0]],[0,jx[1]])
plt.annotate('jx',jx)
```

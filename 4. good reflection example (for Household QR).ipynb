```python
import numpy as np
import matplotlib.pyplot as plt
import sparse_matrix_lib

from mpl_toolkits.mplot3d import Axes3D
```

```python
x = np.array([2,3])
fx = np.array([np.linalg.norm(x), 0])
v = x- fx
v =v/np.linalg.norm(v)
test = x - 2*v*np.dot(v,x)
test
```

```python
x = np.array([2,3])
fx = np.array([np.linalg.norm(x), 0])
v = x + fx
v =v/np.linalg.norm(v)
test = x - 2*v*np.dot(v,x)
test
plt.plot([0,x[0]],[0,x[1]],label = 'x')
plt.annotate('x', x)
plt.plot([0,fx[0]],[0,fx[1]],label = 'fx')
plt.annotate('fx',fx)
plt.plot([0,test[0]],[0,test[1]],label = 'test')
plt.annotate('test', test)
plt.title('Phep quay tot')
```
